```python
# finding null values
movies.isnull().sum()
```

```python
#only three missing values so we simply drop those rows
movies.dropna(inplace=True)

# finding null values again
movies.isnull().sum()
```

```python
# check for any duplicates
movies.duplicated().sum()
```

```python
# genere column is in wiered format
movies.iloc[0].genres
```

```python
# we will try to extract and create a list 
#for ex - ['Action','Adventure','Fantasy','Science Friction']
```

```python
# first we need to convert this string of list to list
# for that we have ast.literal_eval module in python

import ast
```

```python
# creating a helper function

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
# applying helper function on generes
movies['genres'] = movies['genres'].apply(convert)

#updated dataframe after applying helper function on genres
movies.head(1)
```

```python
# applying helper function on keyword
movies['keywords'] = movies['keywords'].apply(convert)

#updated dataframe after applying helper function on keyword
movies.head(1)
```

```python
# in castr column  first 3 actor names are sufficient
# so creating another helper function for that

def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
# applying helper function on cast
movies['cast'] = movies['cast'].apply(convert3)

#updated dataframe after applying helper function on cast
movies.head(1)
```

```python
# we only want to extract director name from the crew from all these dictionaries
movies['crew'][0]
```

```python
#creating a helper function
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
# applying helper function on crew
movies['crew'] = movies['crew'].apply(fetch_director)

#updated dataframe after applying helper function on crew
movies.head(1)
```

```python
# overview colum is a string so we will conver it into list as well so we can concatenate it with other lists
movies['overview'][0]
```

```python
# creating a list of each word
movies['overview'] = movies['overview'].apply(lambda x:x.split())

#updated dataframe
movies.head(1) 
```

```python
# Before concatenating we need to remove an space between words like instead of Sam Worthington we will make it SamWorthington so that Sam and Worthington remains one tag

# creating a helper function
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# applying helper function to columns
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)

#updated dataframe
movies.head(1) 
```

```python
# now we will create a new column namely tags and concatenate all other columns together
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# creating new dataframe that only have movie_id,title and tags column in it
new_df = movies[['movie_id','title','tags']]

#new dataframe
new_df.head(1)
```

```python
# now we will convert the tags list into string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))

#updated dataframe
new_df.head()
```

```python
# now we will convert the tags srting into lower case
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())

#updated dataframe
new_df.head()
```

```python
# importing natural language processing library
import nltk

#importing stemmer which is used to remove morphological affixes from words, leaving only the word stem.and create a object ps from it
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#creting a helper function for stemming tags
def stem(text):
   y = []

   for i in text.split():
     y.append(ps.stem(i))

   return ' '.join(y)  
```

```python
#applying helper function on tags
new_df['tags'] = new_df['tags'].apply(stem)

# updated dataframe
new_df.head(1)
```

```python
# now our problem statement is to recommend 5 most similar movies based on user's interest
# this will be done by the help of tags
# each tags row suggest the name of a movie
# so to find the most similar 5 movies we need to find similarities between the tags
# for that we can try finding sililarity betwwen each tag for recommendation but it is not a very good practice to do
# So instead we will use Text Vectorization and convert the text to vectors and find nearest vectors of each vector for recommendation
# and under Text Vectorization we will be using Bag of words technique
```

```python
# we will use countvectorizer class and use 2 hyperparameters,where stop_words are english words which are used for sectence formation
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# train and convert into numpy array
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
# so we have 4806 movies and 5000 words
#storing it under a new variable

vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
# we can see most commonly used 5000 words
cv.get_feature_names_out()
```

```python
# now we will calculate cosine distance between each 4806 movies
# importing
from sklearn.metrics.pairwise import cosine_similarity

# since 4806 movies so dist between then gives a (4806,4806)shape
cosine_similarity(vector).shape
```

```python
# we will store all distances which we are calculating inside a variable similarity
similarity = cosine_similarity(vector)
```

```python
# we can check the similarity score of movies as well
#first movie has similarity score of 1 with itself and its less for other movies
similarity[0]
```

```python
# now we will create a function where if any movie is given,it will give back 5 similar movies to it
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]#to fetch movie index
    distances = similarity[index]
    distances = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])# find distance and sort in descending order using enumerate function
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
#Cheacking if function works properly
recommend('Avatar')
```

```python
#save files
import pickle

pickle.dump(new_df,open('movie_list.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
